```python
results = []

for col in df2_categorical.columns:
    num_numeric = pd.to_numeric(df2_categorical[col], errors='coerce').notnull().sum()

    # Total count of values in the column
    total_count = len(df2_categorical[col])

    # Count of non-numeric values
    num_non_numeric = total_count - num_numeric

    # Append the results to the list
    results.append([col, num_numeric, num_non_numeric])

result_df = pd.DataFrame(results, columns=['Column Name', 'Numeric Count', 'Non-Numeric Count'])
print(result_df)
```

```python
def has_numeric(row):
    return any(pd.to_numeric(row, errors='coerce').notnull())

# Apply the function row-wise and get boolean mask
mask = df2_categorical.apply(has_numeric, axis=1)

# Drop rows where any element is numeric
df_without_numeric_rows = df2_categorical[~mask]
```
